# 處理 outliers
* 新增欄位註記
* outliers 或 NA 填補
    1. 平均數 (mean)
    2. 中位數 (median, or Q50)
    3. 最大/最小值 (max/min, Q100, Q0)
    4. 分位數 (quantile)

```r
# 設定 data_path
setwd("E:/Users/鄭柏彥/Documents/GitHub/ML-100-Days/data/houseprice")
#讀入檔案
house_train <- read.csv("train.csv", header = TRUE, sep = ",")
```

## 1. 列出 LotFrontage 的 q0 - q100
## 2.1 將 LotFrontage 中的 NA 暫時以中位數填補
## 2.2 將 LotFrontage 的數值標準化至 -1 ~ 1 間
## 3. 將 LotFrontage 的 NA 以眾數填補


```r

quantile(house_train$LotFrontage, 0, na.rm=TRUE)
quantile(house_train$LotFrontage, 1, na.rm=TRUE)
# 1: 計算 LotFrontage 的 q0 - q100
```

```r
# 2.1 將 NAs 以 q50 填補
sum(complete.cases(house_train$LotFrontage)) #完整的資料個數 
sum(is.na(house_train$LotFrontage)) #缺失值總數
which(is.na(house_train$LotFrontage)) #缺失在哪裡
quantile(house_train$LotFrontage, 0.5, na.rm=TRUE)
house_train <- cbind(house_train, house_train$LotFrontage)
house_train[is.na(house_train$LotFrontage), "LotFrontage"] <- median(house_train$LotFrontage, na.rm = TRUE)
```

### Hints: Normalize function (to -1 ~ 1)
$ y = 2*(\frac{x - min(x)}{max(x) - min(x)} - 0.5) $

```r
# 2.2 Normalize values to -1 to 1
normLotFrontage <- 2 * ((house_train$LotFrontage - quantile(house_train$LotFrontage, 0, na.rm=TRUE)) / (quantile(house_train$LotFrontage, 1, na.rm=TRUE) - quantile(house_train$LotFrontage, 0, na.rm=TRUE)) -0.5)
normLotFrontage
```

```r
# 3 列出重複最多的數值
# 設定 data_path
setwd("E:/Users/鄭柏彥/Documents/GitHub/ML-100-Days/data/houseprice")
#讀入檔案
house_train <- read.csv("train.csv", header = TRUE, sep = ",")
which.max(table(house_train$LotFrontage))
house_train[is.na(house_train$LotFrontage), "LotFrontage"] <- names(table(house_train$LotFrontage)[which.max(table(house_train$LotFrontage))])
house_train
```
