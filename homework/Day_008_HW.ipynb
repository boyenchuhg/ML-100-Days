## 作業
1. 將 house_train 中的 SalePrice 依照下列規則分為四組，並將其結果在原本的 dataframe 命名為 CNT_SalePrice_GROUP
    * 價格低於130000 180000 210000
    * 價格低於180000，高於130000
    * 價格低於210000，高於180000
    * 價格高於210000

2. 請根據 CNT_SalePrice_GROUP，列出各組的平均 AMT_LotArea_TOTAL，並繪製 baxplot
3. 請根據 CNT_SalePrice_GROUP，對 AMT_LotArea_TOTAL 計算 [Z 轉換](https://en.wikipedia.org/wiki/Standard_score) 後的分數

```r
#1 分組
CNT_SalePrice_GROUP1 <- subset(house_train, house_train$SalePrice <= 130000)
CNT_SalePrice_GROUP2 <- subset(house_train, house_train$SalePrice > 130000 &  house_train$SalePrice <= 180000)
CNT_SalePrice_GROUP3 <- subset(house_train, house_train$SalePrice > 180000 &  house_train$SalePrice <= 210000)
CNT_SalePrice_GROUP4 <- subset(house_train, house_train$SalePrice > 210000)
```

```r
#2-1 佔地面積平均
AMT_LotArea_TOTAL1 <- mean(CNT_SalePrice_GROUP1$LotArea)
AMT_LotArea_TOTAL2 <- mean(CNT_SalePrice_GROUP2$LotArea)
AMT_LotArea_TOTAL3 <- mean(CNT_SalePrice_GROUP3$LotArea)
AMT_LotArea_TOTAL4 <- mean(CNT_SalePrice_GROUP4$LotArea)
AMT_LotArea_TOTAL1 
AMT_LotArea_TOTAL2 
AMT_LotArea_TOTAL3 
AMT_LotArea_TOTAL4 
```

```r
#2-2 boxplot
boxplot(CNT_SalePrice_GROUP1$LotArea, CNT_SalePrice_GROUP2$LotArea, CNT_SalePrice_GROUP3$LotArea, CNT_SalePrice_GROUP4$LotArea)
```

```r
#3 Z轉換(根據不同分組)
scale(CNT_SalePrice_GROUP1$LotArea, center = TRUE, scale = TRUE)
scale(CNT_SalePrice_GROUP2$LotArea, center = TRUE, scale = TRUE)
scale(CNT_SalePrice_GROUP3$LotArea, center = TRUE, scale = TRUE)
scale(CNT_SalePrice_GROUP4$LotArea, center = TRUE, scale = TRUE)
```
