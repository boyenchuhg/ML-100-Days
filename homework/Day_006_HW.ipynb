# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```r
# 設定 data_path
setwd("E:/Users/鄭柏彥/Documents/GitHub/ML-100-Days/data/houseprice")
#讀入檔案
house_train <- read.csv("train.csv", header = TRUE, sep = ",")
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

使用資料為房地產資料，許多為類型，無離群值，僅有
1. LotArea: Lot size in square feet
2. 1stFlrSF: First Floor square feet
3. GrLivArea: Above grade (ground) living area square feet
4. GarageArea: Size of garage in square feet
5. SalePrice
可能具有離群值

```r
summary(house_train$LotArea)
x1 <- seq(min(house_train$LotArea),max(house_train$LotArea),length.out=1460)
mean1 <- mean(house_train$LotArea)
SD1 <- sd(house_train$LotArea)
plot.ecdf(x = house_train$LotArea) #超過5萬平方英尺就需要注意一下，不一定不合理，但是必然是特別的物件
lines(x1, pnorm(x1,mean1,SD1),col="green")
```

```r
summary(house_train$lstFlrSF)
x2 <- seq(min(house_train$lstFlrSF),max(house_train$lstFlrSF),length.out=1460)
mean2 <- mean(house_train$lstFlrSF)
SD2 <- sd(house_train$lstFlrSF)
plot.ecdf(x = house_train$lstFlrSF) #超過兩千？
lines(x2, pnorm(x2,mean2,SD2),col="red")
```

```r
summary(house_train$GrLivArea)
x3 <- seq(min(house_train$GrLivArea),max(house_train$GrLivArea),length.out=1460)
mean3 <- mean(house_train$GrLivArea)
SD3 <- sd(house_train$GrLivArea)
plot.ecdf(x = house_train$GrLivArea) #超過三千？
lines(x3, pnorm(x3,mean3,SD3),col="red")
```

```r
summary(house_train$GarageArea)
x4 <- seq(min(house_train$GarageArea),max(house_train$GarageArea),length.out=1460)
mean4 <- mean(house_train$GarageArea)
SD4 <- sd(house_train$GarageArea)
plot.ecdf(x = house_train$GarageArea) #超過一千？
lines(x4, pnorm(x4,mean4,SD4),col="red")
```

```r
summary(house_train$SalePrice)
x5 <- seq(min(house_train$SalePrice),max(house_train$SalePrice),length.out=1460)
mean5 <- mean(house_train$SalePrice)
SD5 <- sd(house_train$SalePrice)
plot.ecdf(x = house_train$SalePrice) #超過4*10^5？
lines(x5, pnorm(x5,mean5,SD5),col="red")
```
