```r
# 設定 data_path
setwd("E:/Users/鄭柏彥/Documents/GitHub/ML-100-Days/data/houseprice")
#讀入檔案
house_train <- read.csv("train.csv", header = TRUE, sep = ",")
```

無監督的數據離散化

```r
#等寬分箱
A1 <- discretize(house_train$SalePrice,"equalwidth",5)
table(A1)
plot(house_train$SalePrice,col=A1$X)
```

```r
#等頻率分箱
A2 <- discretize(house_train$SalePrice,"equalfreq",5)
table(A2)
plot(house_train$SalePrice,col=A2$X)
```

```r
#kmeans分箱
A3 <- kmeans(house_train$SalePrice, 5)
table(A3$cluster)
A3$centers
plot(house_train$SalePrice,col=A3$cluster)
```
